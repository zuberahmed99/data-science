```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv("PRSA_data_2010.1.1-2014.12.31.csv")
data.head()
```

```python
data.dropna(axis=0,inplace=True)
```

```python
data.head()
data.describe()
```

```python
sns.pairplot(data)
```

```python
def create_date(cols):
    year = cols[0]
    month = cols[1]
    day = cols[2]
    
    if month<10:
        month = "0" + str(month)
        
    if day<10:
        day = "0" + str(day)
    date = str(year) + "-" + str(month) +"-" + str(day)
    return date

    
```

```python
data['date'] = data[['year','month','day']].apply(create_date, axis = 1)
data.head()
```

```python
def convert_to_kelvin(temp):
    kelvin = temp + 273
    return kelvin

data['TEMP'] = data['TEMP'].apply(convert_to_kelvin)
data.head()
    
```

```python
df = data[['date','month','day','hour','pm2.5','TEMP','PRES']]
```

```python
df.head()
```

```python
#sampling a small part of the data to find patterns

df1 = df.head(1000)
```

```python
df1.describe()
pm_by_hour = df1['pm2.5'].groupby(df1['hour']).mean()
```

```python
type(pm_by_hour)
pm_by_hour.head()
```

```python
plt.plot(pm_by_hour)
plt.show()
```

#### It shows that during midnight the pm2.5 levels are high.
